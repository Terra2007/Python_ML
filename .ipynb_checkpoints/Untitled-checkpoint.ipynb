```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
score = [53., 57., 61., 63., 66., 67., 69.]
hour = [1., 2., 3., 4., 5., 6., 7.]
study = pd.DataFrame({"Score" : score,
                      "Hour" : hour})
```

```python
study.head(7)
```

```python
sns.lmplot(x = 'Hour', y = 'Score', data = study)
```
